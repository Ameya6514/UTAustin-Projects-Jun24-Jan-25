# Step 1: Load Libraries and Dataset


```python
# Importing essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
data = pd.read_csv('austo_automobile.csv')

# Display basic information about the dataset
data.info()
data.head()
```

# Step 2: Data Understanding and Cleaning

```python
# Checking for missing values
print(data.isnull().sum())
```

```python
# Dropping or imputing missing values based on observations
data = data.dropna()  # Alternatively, impute values if relevant
```

```python
# Convert categorical variables if needed
categorical_columns = ['Gender', 'Profession', 'Marital_status', 'Education', 'Make']
for col in categorical_columns:
    data[col] = data[col].astype('category')
```

# Step 3: Univariate Data Analysis

```python
# Univariate analysis of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=15, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.show()
```

```python
# Univariate analysis of Gender
plt.figure(figsize=(6, 4))
sns.countplot(data['Gender'])
plt.title('Distribution of Gender')
plt.show()
```

```python
# Univariate Analysis for Profession
plt.figure(figsize=(6, 4))
sns.countplot(data['Profession'])
plt.title('Profession Distribution')
plt.show()
```

```python
# Univariate Analysis for Marital_status
plt.figure(figsize=(6, 4))
sns.countplot(data['Marital_status'])
plt.title('Marital Status Distribution')
plt.show()
```

```python
# Univariate Analysis for Education
plt.figure(figsize=(8, 5))
sns.countplot(data['Education'])
plt.title('Education Level Distribution')
plt.show()
```

```python
# Univariate Analysis for Partner_salary
plt.figure(figsize=(10, 6))
sns.histplot(data['Partner_salary'], bins=20, kde=True, color='green')
plt.title('Partner Salary Distribution')
plt.xlabel('Partner Annual Salary')
plt.show()
```

```python
# Univariate Analysis for Price
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], bins=20, kde=True, color='red')
plt.title('Car Price Distribution')
plt.xlabel('Price')
plt.show()
```

# Step 4: Multivariate Data Analysis

```python
# Bivariate analysis to explore relationships between pairs of variables.
# Correlation analysis for numerical variables like Salary, Partner_salary, Total_salary, and Price.
# Categorical vs Numerical Analysis to understand how customer characteristics vary by car type.
```

```python
# Bivariate analysis - Age vs Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Price', hue='Make', data=data)
plt.title('Age vs Price by Car Type')
plt.show()
```

```python
# Correlation heatmap for numerical variables
plt.figure(figsize=(10, 6))
sns.heatmap(data[['Age', 'Salary', 'Partner_salary', 'Total_salary', 'Price']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Analyzing Salary by Car Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Make', y='Salary', data=data)
plt.title('Salary Distribution by Car Type')
plt.show()
```
